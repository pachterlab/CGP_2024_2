```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
gehring = pd.read_csv('gehring_corrs_df.csv')
gehring
```

```python
norman = pd.read_csv('norman_corrs_df.csv')
norman
```

```python
replogle = pd.read_csv('replogle_corrs_df.csv')
replogle
```

```python
gehring['Dataset'] = ['Gehring']*len(gehring)
norman['Dataset'] = ['Norman']*len(norman)
replogle['Dataset'] = ['Replogle']*len(replogle)
```

```python
all_dfs = pd.concat([gehring,norman,replogle])
```

```python
all_dfs
```

```python
def conv_cond(x):
    '''
    Convert conditions into which genes used (all, RF,logreg)
    '''
    
    if 'all' in x:
        return 'all'
    elif 'RF' in x:
        return 'RF'
    else:
        return 'logreg'
```

```python
all_dfs['Genes Used'] =  [conv_cond(x) for x in all_dfs.Cond]
print(all_dfs.head())
```

```python
col_b = ['#1B9AAA']
col_beta = ['#9fcbee']
col_gamma =['#3C50B1']
col_expr = ['#9299A0']

palette={'b':col_b[0],'beta':col_beta[0],'gamma':col_gamma[0],'expr':col_expr[0]}
```

```python
all_dfs['Data_Model'] = [i+'_'+j for i,j in zip(list(all_dfs['Dataset']),list(all_dfs['Pred Model']))]
all_dfs.head()
```

```python
only_conds = list(all_dfs['Cond'])
only_conds = [i.replace('all','') for i in only_conds]
only_conds = [i.replace('RF','') for i in only_conds]
only_conds = [i.replace('logreg','') for i in only_conds]

all_dfs['Only_Cond'] = only_conds
```

```python
cond_filt = all_dfs['Genes Used'].isin(['RF'])
model_filt = all_dfs['Pred Model'].isin(['Mult','Add'])
param_filt_b = all_dfs['Param'].isin(['b'])
param_filt_beta = all_dfs['Param'].isin(['beta'])

sub_dfs_b = all_dfs[cond_filt&model_filt&param_filt_b]
sub_dfs_beta = all_dfs[cond_filt&model_filt&param_filt_beta]

plt.figure(figsize=(8,3.5))
sns.stripplot(data=sub_dfs_b,x='Data_Model',y='Corr',hue='Only_Cond',
              palette='Set3',dodge=True,size=9,edgecolor='black',linewidth=0.5,
               jitter=0.05,marker='o',alpha=0.7)
g=sns.stripplot(data=sub_dfs_beta,x='Data_Model',y='Corr',hue='Only_Cond',
              palette='Set3',dodge=True,size=9,edgecolor='black',linewidth=0.5,
               jitter=0.1,marker='^',alpha=0.7)

sns.move_legend(g, "upper left", bbox_to_anchor=(1, 1),ncol=2)
plt.xticks(rotation=90)

plt.ylim([-0.4,1.0])
plt.tight_layout()
plt.savefig('alldata_pred_corrs_multvadd.pdf')
```

```python
cond_filt = all_dfs['Genes Used'].isin(['RF'])
model_filt = all_dfs['Pred Model'].isin(['Mult','Mult Neg'])
param_filt_b = all_dfs['Param'].isin(['b'])
param_filt_beta = all_dfs['Param'].isin(['beta'])

sub_dfs_b = all_dfs[cond_filt&model_filt&param_filt_b]
sub_dfs_beta = all_dfs[cond_filt&model_filt&param_filt_beta]

plt.figure(figsize=(8,3.5))
sns.stripplot(data=sub_dfs_b,x='Data_Model',y='Corr',hue='Only_Cond',
              palette='Set3',dodge=True,size=7,edgecolor='black',linewidth=0.5,
               jitter=0.05,marker='o',
              order=['Gehring_Mult Neg','Gehring_Mult','Norman_Mult Neg','Norman_Mult',
                                 'Replogle_Mult Neg','Replogle_Mult'])
g=sns.stripplot(data=sub_dfs_beta,x='Data_Model',y='Corr',hue='Only_Cond',
              palette='Set3',dodge=True,size=7,edgecolor='black',linewidth=0.5,
               jitter=0.05,marker='^',
               order=['Gehring_Mult Neg','Gehring_Mult','Norman_Mult Neg','Norman_Mult',
                                 'Replogle_Mult Neg','Replogle_Mult'])

sns.move_legend(g, "upper left", bbox_to_anchor=(1, 1),ncol=2)
plt.xticks(rotation=90)

plt.ylim([-0.4,1.0])
plt.tight_layout()
plt.savefig('alldata_pred_corrs_multvneg.pdf')
```
